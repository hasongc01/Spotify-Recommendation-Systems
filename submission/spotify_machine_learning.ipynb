### Decision Tree, Random Forest

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import re
from sklearn.model_selection import train_test_split
#df = df[(df['year'] >= 2015) & (df['year'] < 2020)]
df = df[(df['year'] == 2019) ]
percentile_75 = df['track_popularity'].quantile(0.75)
percentile_85 = df['track_popularity'].quantile(0.85)
percentile_95 = df['track_popularity'].quantile(0.95)
df['popularity_75'] = (df['track_popularity'] >= percentile_75).astype(int)
df['popularity_85'] = (df['track_popularity'] >= percentile_85).astype(int)
df['popularity_95'] = (df['track_popularity'] >= percentile_95).astype(int)
corr_feature1 = df['energy'].corr(df['track_popularity'])
corr_feature2 = df['loudness'].corr(df['track_popularity'])
total_corr = abs(corr_feature1) + abs(corr_feature2)
weight1 = abs(corr_feature1) / total_corr
weight2 = abs(corr_feature2) / total_corr

df['energy_loud'] = weight1 * df['energy'] + weight2 * df['loudness']
new_df = df.drop('track_popularity', axis = 1)
new_df['energy_loud'] = new_df['energy'] + new_df['loudness']

new_df['mode'] = [str(x) for x in new_df['mode']]
new_df['key'] = [str(x) for x in new_df['key']]
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Transform the duration column
new_df['duration_normalized'] = scaler.fit_transform(new_df[['duration_ms']])
new_df = new_df[['danceability', 'energy_loud', 'key', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_normalized', 'popularity_75', 'popularity_85', 'popularity_95']]
columns_to_clean = ['key', 'mode']

for col in columns_to_clean:
    new_df[col] = new_df[col].str.replace(r'\s+', '', regex=True)  # Remove all spaces
vectorized_data = pd.get_dummies(new_df[['key', 'mode']], drop_first=False)
new_df = pd.concat([new_df, vectorized_data], axis=1)
new_df = new_df.drop(columns_to_clean, axis = 1)
new_df
```

```python
def decision_tree(y):
    test_size, random_state = 0.3, 42
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    from sklearn.model_selection import  RandomizedSearchCV
    from sklearn.tree import DecisionTreeClassifier
    from scipy.stats import uniform, poisson
    import numpy as np

    parameters = {
        'splitter': ['best', 'random'],   # Splitting strategy
        'class_weight': [None, 'balanced'],  # Class weighting for imbalance
        'max_depth': [3, 6, 9, None],  # Depth of the tree
        'max_leaf_nodes': [3, 6, 9, None],  # Maximum number of leaf nodes
        'criterion': ['gini', 'entropy', 'log_loss', None],  # Splitting criterion
        'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
        'min_samples_leaf': [1, 2, 4], # Minimum number of samples required to be at a leaf node
        #'ccp_alpha': uniform(0.0, 0.05) 
    }

    r_dt = RandomizedSearchCV(
        DecisionTreeClassifier(), 
        parameters, 
        cv=20, 
        n_iter=100, 
        random_state=42, 
        n_jobs=-1,
        scoring = 'f1'
    )
    r_dt.fit(X_train,y_train)
    cls_params = r_dt.best_params_
    print(cls_params)
    print()
    from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

    model = r_dt.best_estimator_
    predicted = model.predict(X_train)
    # Calculate performance metrics
    accuracy = accuracy_score(y_train, predicted)
    precision = precision_score(y_train, predicted)
    recall = recall_score(y_train, predicted)
    f1 = f1_score(y_train, predicted)
    conf_matrix = confusion_matrix(y_train, predicted)
    predicted_train = predicted
    # Print the performance metrics
    print("For train set")
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1 Score: {f1}')
    print(conf_matrix)
    predicted = model.predict(X_test)
    # Calculate performance metrics
    accuracy = accuracy_score(y_test, predicted)
    precision = precision_score(y_test, predicted)
    recall = recall_score(y_test, predicted)
    f1 = f1_score(y_test, predicted)
    conf_matrix = confusion_matrix(y_test, predicted)
    predicted_test = predicted
    # Print the performance metrics
    print()
    print("For test set")
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1 Score: {f1}')
    print(conf_matrix)
    print()
    feature_importances = model.feature_importances_

    features = X_train.columns
    feature_importance_df = pd.DataFrame({
        'Feature': features,
        'Importance': feature_importances
    }).sort_values(by='Importance', ascending=False)

    print(feature_importance_df)
```

```python
print("at 0.75 level")
y = new_df['popularity_75']
decision_tree(y)
print("at 0.85 level")
y = new_df['popularity_85']
decision_tree(y)
print("at 0.95 level")
y = new_df['popularity_95']
decision_tree(y)
```

```python
def random_forest(y):
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.model_selection import  RandomizedSearchCV
    test_size, random_state = 0.3, 42
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    # Create the param grid
    param_grid = { 
        'n_estimators': [5, 25, 50, 100, 150],  # Number of trees
        'max_features': ['sqrt', 'log2', None],  # Number of features to consider for splits
        'max_depth': [3, 6, 9, None],  # Depth of the tree
        'max_leaf_nodes': [3, 6, 9, None],  # Maximum number of leaf nodes
        'criterion': ['gini', 'entropy', 'log_loss'],  # Splitting criterion
        'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
        'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
        'class_weight': [None, 'balanced']  # Class weighting for imbalanc
    }
    rf_Model = RandomForestClassifier(random_state=42)
    r_rf = RandomizedSearchCV(estimator = rf_Model, param_distributions = param_grid, cv = 10, verbose=2, n_jobs = 4, scoring='f1')
    r_rf.fit(X_train, y_train)
    print(r_rf.best_params_)
    print()
    model = r_rf.best_estimator_
    predicted = model.predict(X_train)
    from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

    # Calculate performance metrics
    accuracy = accuracy_score(y_train, predicted)
    precision = precision_score(y_train, predicted)
    recall = recall_score(y_train, predicted)
    f1 = f1_score(y_train, predicted)
    conf_matrix = confusion_matrix(y_train, predicted)
    predicted_train = predicted
    # Print the performance metrics
    print("For train set")
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1 Score: {f1}')
    print(conf_matrix)

    predicted = model.predict(X_test)
    # Calculate performance metrics
    accuracy = accuracy_score(y_test, predicted)
    precision = precision_score(y_test, predicted)
    recall = recall_score(y_test, predicted)
    f1 = f1_score(y_test, predicted)
    conf_matrix = confusion_matrix(y_test, predicted)
    predicted_test = predicted
    # Print the performance metrics
    print()
    print("For test set")
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1 Score: {f1}')
    print(conf_matrix)
    print()
    feature_importances = model.feature_importances_

    features = X_train.columns
    feature_importance_df = pd.DataFrame({
        'Feature': features,
        'Importance': feature_importances
    }).sort_values(by='Importance', ascending=False)

    print(feature_importance_df)
```

### Logistic Regression

Binary Classification of Track Popularity

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
q75 = df['track_popularity'].quantile(0.75)
q85 = df['track_popularity'].quantile(0.85)
q95 = df['track_popularity'].quantile(0.95)

df['popularity_75'] = 0 
df.loc[df['track_popularity'] > q75, 'popularity_75'] = 1  

df['popularity_85'] = 0 
df.loc[df['track_popularity'] > q85, 'popularity_85'] = 1  

df['popularity_95'] = 0 
df.loc[df['track_popularity'] > q95, 'popularity_95'] = 1  

df
```

```python
numerical_features = ['danceability','energy_loud','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']
categorical_features = ['mode', 'key']
```

```python
df_encoded_categorical = pd.get_dummies(df[categorical_features], drop_first=True)
df_encoded = pd.concat([df[numerical_features], df_encoded_categorical], axis=1)
```

Q75

```python
y_75 = df['popularity_75']
X = df_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_75, test_size=0.3, random_state=42)


model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# using x_train
predicted = model.predict(X_train)

# performance metrics
accuracy = accuracy_score(y_train, predicted)
precision = precision_score(y_train, predicted)
recall = recall_score(y_train, predicted)
f1 = f1_score(y_train, predicted)
conf_matrix = confusion_matrix(y_train, predicted)
predicted_train = predicted

# print the performance metrics
print("For train set (q75)")
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(conf_matrix)
print()

# using x_test
y_prediction = model.predict(X_test)

# performance metrics
accuracy = accuracy_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction)
recall = recall_score(y_test, y_prediction)
f1 = f1_score(y_test, y_prediction)
conf_matrix = confusion_matrix(y_test, y_prediction)

# print the performance metrics
print("For test set (q75)")
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print('Confusion Matrix:')
print(conf_matrix)
```

```python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
}).sort_values(by='Coefficient', ascending=False)

print(coefficients)
```
