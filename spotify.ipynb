# Data Visualization

### Word Cloud

```python
#pip install wordcloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df['track_album_release_date'] = df['track_album_release_date'].astype(str).str.strip()
df['year'] = df['track_album_release_date'].str[:4]
df['year'] = pd.to_numeric(df['year'], errors='coerce')
years = [2010, 2015, 2020]
```

```python
# avoid including version information
df['track_name_cleaned'] = df['track_name'].str.split('-').str[0].str.strip()
df['track_name_cleaned'] = df['track_name_cleaned'].str.split('feat', expand=True)[0].str.strip()
df['track_name_cleaned'] = df['track_name_cleaned'].str.split('Remix', expand=True)[0].str.strip()
filtered_data = df[df['year'].isin(years)]
title_text_cleaned = {year: ' '.join(filtered_data[filtered_data['year'] == year]['track_name_cleaned'].dropna().astype(str)) for year in years}
artist_text = {year: ' '.join(filtered_data[filtered_data['year'] == year]['track_artist'].dropna().astype(str)) for year in years}
plt.figure(figsize=(15, 8))
for i, year in enumerate(years):
    plt.subplot(1, 3, i + 1)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(title_text_cleaned[year])
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"Most Common Song Titles in {year}", fontsize=16)
    plt.axis('off')

plt.tight_layout()
plt.show()
plt.figure(figsize=(15, 8))
for i, year in enumerate(years):
    plt.subplot(1, 3, i + 1)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(artist_text[year])
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"Most Common Artist Names in {year}", fontsize=16)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
