### Decision Tree, Random Forest

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import re
from sklearn.model_selection import train_test_split
mean_popularity = df['track_popularity'].mean()
df['popularity'] = (df['track_popularity'] >= mean_popularity).astype(int)
new_df = df.drop('track_popularity', axis = 1)
new_df = new_df[(new_df['year'] >= 2015) & (new_df['year'] < 2020)]
new_df = new_df[['track_artist', 'playlist_genre', 'playlist_subgenre', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'popularity']]

columns_to_clean = ['track_artist', 'playlist_genre', 'playlist_subgenre']

for col in columns_to_clean:
    new_df[col] = new_df[col].str.replace(r'\s+', '', regex=True)  # Remove all spaces
vectorized_data = pd.get_dummies(new_df[['playlist_genre', 'playlist_subgenre']], drop_first=False)
new_df = pd.concat([new_df, vectorized_data], axis=1)
new_df = new_df.drop(columns_to_clean, axis = 1)
new_df
```

```python
X = new_df.drop('popularity', axis=1)
X = X
y = new_df['popularity']
```

```python
test_size, random_state = 0.3, 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python
X_train
```

```python
from sklearn.model_selection import  RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from scipy.stats import uniform, poisson
import numpy as np

parameters = {
    'criterion': ['gini', 'entropy'],  # Measure for splits
    'splitter': ['best', 'random'],   # Splitting strategy
    'max_depth': poisson(mu=3, loc=2),  # Maximum depth of the tree
    'min_samples_split': uniform(0.01, 0.2),  # Fraction of samples for a split
    'min_samples_leaf': poisson(mu=1, loc=1),  # Minimum samples at a leaf node
    'max_leaf_nodes': poisson(mu=4, loc=3),  # Maximum leaf nodes
    'class_weight': [None, 'balanced'],  # Class weighting for imbalance
    'ccp_alpha': uniform(0.0, 0.05)  # Pruning parameter
}

r_dt = RandomizedSearchCV(
    DecisionTreeClassifier(), 
    parameters, 
    cv=10, 
    n_iter=100, 
    random_state=42, 
    n_jobs=-1, 
    scoring='f1'
)
r_dt.fit(X_train,y_train)
cls_params = r_dt.best_params_
cls_params['min_samples_split'] = np.ceil(cls_params['min_samples_split']*X_train.shape[0])
cls_params
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

model = r_dt.best_estimator_
predicted = model.predict(X_train)
# Calculate performance metrics
accuracy = accuracy_score(y_train, predicted)
precision = precision_score(y_train, predicted)
recall = recall_score(y_train, predicted)
f1 = f1_score(y_train, predicted)
conf_matrix = confusion_matrix(y_train, predicted)
predicted_train = predicted
# Print the performance metrics
print("For train set")
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(conf_matrix)
predicted = model.predict(X_test)
# Calculate performance metrics
accuracy = accuracy_score(y_test, predicted)
precision = precision_score(y_test, predicted)
recall = recall_score(y_test, predicted)
f1 = f1_score(y_test, predicted)
conf_matrix = confusion_matrix(y_test, predicted)
predicted_test = predicted
# Print the performance metrics
print()
print("For test set")
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(conf_matrix)
```

```python
feature_importances = model.feature_importances_

features = X_train.columns
feature_importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print(feature_importance_df)
```

### Linear Regression

Initial Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
df['track_album_release_date'] = df['track_album_release_date'].astype(str).str.strip()
df['year'] = df['track_album_release_date'].str[:4]
df['year'] = pd.to_numeric(df['year'], errors='coerce')

df = df[(df['year'] >= 2015) & (df['year'] <= 2019)]
```

```python
# combine features with high correlation
df['loudness_energy'] = df['loudness'] + df['energy']
```

```python
numerical_features = ['danceability','loudness_energy','key','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']
categorical_features = ['playlist_genre', 'playlist_subgenre', 'year']
```

```python
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

df_encoded_categorical = pd.get_dummies(df[categorical_features], drop_first=True)
df_encoded = pd.concat([df[numerical_features], df_encoded_categorical], axis=1)
```

```python
y = df['track_popularity']

# encode the categorical features
X = df_encoded
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_prediction = model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_prediction)
mse = mean_squared_error(y_test, y_prediction)

coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
}).sort_values(by='Coefficient', ascending=False)

print(f'R-Squared: {r2}')
print(f'Mean Squared Error: {mse}')
print(coefficients)
```

Top Features and Re-Regression Based on Significant Coefficients

```python
coefficients_sorted = coefficients.copy()
coefficients_sorted['Abs_Coefficient'] = coefficients_sorted['Coefficient'].abs()
coefficients_sorted = coefficients_sorted.sort_values(by='Abs_Coefficient', ascending=False)
coefficients_sorted.head(10)
```

```python
top_features = coefficients_sorted['Feature'].head(10).tolist()

X_train_top = X_train[top_features]
X_test_top = X_test[top_features]

model_top = LinearRegression()
model_top.fit(X_train_top, y_train)

y_prediction_top = model_top.predict(X_test_top)
```

```python
r2_top = r2_score(y_test, y_prediction_top)
mse_top = mean_squared_error(y_test, y_prediction_top)

coefficients_top = model_top.coef_
coefficients_top = pd.DataFrame({
    'Feature': top_features,
    'Coefficient': coefficients_top
}).sort_values(by='Coefficient', ascending=False)

print(f'R-Squared (for top 10 features): {r2_top}')
print(f'Mean Squared Error (for top 10 features): {mse_top}')

coefficients_top
```

### Logistic Regression

Binary Classification of Track Popularity

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
q75 = df['track_popularity'].quantile(0.75)
q25 = df['track_popularity'].quantile(0.25)

# Encode popularity
df['popularity'] = 0 
df.loc[df['track_popularity'] > q75, 'popularity'] = 1  

# Drop entries in between the 25th and 75th percentiles
df = df[(df['track_popularity'] > q75) | (df['track_popularity'] < q25)]
```

```python
numerical_features = ['danceability','loudness_energy','key','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']
categorical_features = ['playlist_genre','playlist_subgenre', 'year']
```

```python
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

df_encoded_categorical = pd.get_dummies(df[categorical_features], drop_first=True)
df_encoded = pd.concat([df[numerical_features], df_encoded_categorical], axis=1)
```

```python
y = df['popularity']
X = df_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


model = LogisticRegression()
model.fit(X_train, y_train)

y_prediction = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction)
recall = recall_score(y_test, y_prediction)
f1 = f1_score(y_test, y_prediction)
conf_matrix = confusion_matrix(y_test, y_prediction)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print('Confusion Matrix:')
print(conf_matrix)
```

```python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
}).sort_values(by='Coefficient', ascending=False)

print(coefficients)
```
